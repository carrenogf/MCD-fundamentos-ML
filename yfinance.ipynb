```python
import yfinance as yf
import pandas as pd
import sqlite3
```

```python
features = ["symbol","shortName","sector","country","fullTimeEmployees","auditRisk","overallRisk","dividendYield","payoutRatio","fiveYearAvgDividendYield","beta","forwardPE","averageVolume","marketCap","ebitdaMargins","earningsGrowth","returnOnEquity"]
```

```python
# extraer tickers de S&P 500 desde Wikipedia

# URL de la tabla en Wikipedia
url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"

# Leer las tablas de la página
tables = pd.read_html(url)

# Extraer la primera tabla (que contiene la lista de empresas)
sp500_df = tables[0]

# Obtener la lista de tickers
tickers_sp500 = sp500_df["Symbol"].tolist()

print(tickers_sp500)
```

```python
len(tickers_sp500)
```

```python
conn = sqlite3.connect('sp500.db')
cursor = conn.cursor()
#cursor.execute(""" DROP TABLE IF EXISTS acciones """)
cursor.execute('''
CREATE TABLE IF NOT EXISTS acciones (
  id INTEGER PRIMARY KEY AUTOINCREMENT,
  symbol TEXT NULL,
  shortName TEXT NULL,
  sector TEXT NULL,
  country TEXT NULL,
  fullTimeEmployees REAL NULL,
  auditRisk REAL NULL,
  overallRisk REAL NULL,
  dividendYield REAL NULL,
  payoutRatio REAL NULL,
  fiveYearAvgDividendYield REAL NULL,
  beta REAL NULL,
  forwardPE REAL NULL,
  averageVolume REAL NULL,
  marketCap REAL NULL,
  ebitdaMargins REAL NULL,
  earningsGrowth REAL NULL,
  returnOnEquity REAL NULL)
''')

def guardar_datos(ticker):
    try:
        # Obtener datos de la acción
        ticker_data = yf.Ticker(ticker).info
        # Filtrar solo las características deseadas
        ticker_data = {k: ticker_data[k] for k in features if k in ticker_data}
        # Convertir a DataFrame y guardar en SQLite
        df = pd.DataFrame([ticker_data])
        df.to_sql('acciones', conn, if_exists='append', index=False)
        conn.commit()
    except Exception as e:
        print(f"Error al guardar datos para {ticker}: {e}")
        
```

```python
conn = sqlite3.connect('sp500.db')
for ticker in tickers_sp500:
    try:
        guardar_datos(ticker)
        print("se guardaron los datos de ", ticker)
    except Exception as e:
        print(f"Error al procesar {ticker}: {e}")
        continue
conn.close()
```
